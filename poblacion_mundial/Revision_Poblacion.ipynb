## Poblacion Periodo 2000-2020

```python
datos= pd.read_csv('data/poblacion.csv')
```

```python
d= datos[(datos.Time>1998) & (datos.Time<2021)][['Location','VarID','Variant','Time','PopTotal']]
d
```

```python
d = d.set_index('Location')
```

```python
d
```

```python
dv = d[d.Variant=='Medium'][['VarID','Time','PopTotal']]
dv.head()
```

```python
t = dv[['Time']]
t.head()
```

```python
p = dv[['PopTotal']]
p
```

```python
pob = p.diff(axis=0,periods=1)
pob
```

```python
df2= t.assign(Time=t, PopTotal=p, Inc= pob)
print(df2)
```

```python
e= df2.PopTotal
e.head(3)
```

```python
s =e.shift(1,axis=0)
s
```

```python
df2= t.assign(Time=t, PopTotal=p, Inc= pob, PobAnt=s)
print(df2)
```

```python
tas = round((df2.Inc/df2.PobAnt)*100,2)
```

```python
tas.head(3)
```

```python
df2= t.assign(Time=t, PopTotal=p, Inc= pob, PobAnt=s, Tasa= tas)
print(df2)
```

```python
df2
```

```python
round(df2.groupby('Location')['Tasa'].mean(),2)
```
